```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
import csv
import matplotlib.patches as mpatches
import pylab as pb
import matplotlib.ticker as mtick
import scipy.misc
from mpl_toolkits.mplot3d import Axes3D
from skimage.io import imread
from matplotlib import cm as cm
```

```python

x1 = []
y1 = []
z1 = []
a1 = []
b1 = []
c1 = []
d1 = []
e1 = []


with open('Raman_all.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x1.append(float(row[0]))
        y1.append(float(row[1]))
        z1.append(float(row[2]))
        a1.append(float(row[3]))
        b1.append(float(row[4]))
        c1.append(float(row[5]))
        d1.append(float(row[6]))
        
```

```python

x2 = []
y2 = []
z2 = []
a2 = []
b2 = []
c2 = []
d2 = []
e2 = []


with open('PL_all.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x2.append(float(row[0]))
        y2.append(float(row[1]))
        z2.append(float(row[2]))
        a2.append(float(row[3]))
        b2.append(float(row[4]))
        c2.append(float(row[5]))
        d2.append(float(row[6]))
        
```

```python
fig = plt.figure(figsize=(3,3))

axes = fig.add_subplot(111)

axes.set_xlabel('Wavenumber (cm$^{-1}$)', fontsize = 12)
axes.set_ylabel('Intensity (a.u.)', fontsize = 12)
plt.yticks(fontsize = 12)
plt.xticks(fontsize = 12)

axes.set_yscale("log")
axes.grid('off')
plt.setp(axes.get_yticklabels(), visible=False, fontsize = 12)
y1 = [y1+200 for y1 in y1]

z1 = [z1+100 for z1 in z1]

a1 = [a1+80 for a1 in a1]

b1 = [b1+50 for b1 in b1]

c1 = [c1+25 for c1 in c1]
axes.plot(x1,y1, color='black', linestyle='-', linewidth=1.5, label = '0 s' )
#axes.plot(x1,z1, color='blue', linestyle='-', linewidth=1.5, label = '2 s')
axes.plot(x1,a1, color='g', linestyle='-', linewidth=1.5, label = '4 s')
axes.plot(x1,b1, color='red', linestyle='-', linewidth=1.5, label = '6 s')
axes.plot(x1,c1, color='orange', linestyle='-', linewidth=1.5, label = '8 s')
axes.plot(x1,d1, color='darkviolet', linestyle='-', linewidth=1.5, label = '10 s')
major_ticks_1 = np.arange(340, 480, 20)   
axes.set_xticks(major_ticks_1)  

axes.set_xlim([355,435])
axes.set_ylim([220,720])

plt.setp(axes.get_yticklabels(), visible=False, fontsize = 12)
plt.setp(axes.get_xticklabels(), visible=True, fontsize = 12)
axes.legend(bbox_to_anchor=(0.25, 1.15),labelspacing=0.01, ncol = 1,fontsize = 8)



plt.savefig("4L_Raman.png", bbox_inches='tight',dpi = 300)
plt.show()
```

```python
fig = plt.figure(figsize=(3,3))

axes = fig.add_subplot(111)

axes.set_xlabel('Energy (eV)', fontsize = 12)
axes.set_ylabel('Intensity (a.u.)', fontsize = 12)
plt.yticks(fontsize = 12)
plt.xticks(fontsize = 12)
axes.grid('off')
axes.set_xlim([1.75,2.15])
axes.set_ylim([220,750])
major_ticks_1 = np.arange(1.75, 2.2, 0.1)   
axes.set_xticks(major_ticks_1)  

axes.plot(x2,y2, "-", color='black', lw = 1.5,  label = '0 s',alpha=1  )
#axes.plot(x2,z2,"-", color='blue', lw = 1.5,  label = '2 s',alpha=1 )
axes.plot(x2,a2, "-", color='g',lw = 1.5,   label = '4 s' ,alpha=1)
axes.plot(x2,b2, "-", color='r', lw = 1.5,  label = '6 s' ,alpha=1)
axes.plot(x2,c2, "-", color='orange',lw = 1.5,   label = '8 s' ,alpha=1)
axes.plot(x2,d2, "-", color='darkviolet',lw = 1.5,   label = '10 s' ,alpha=1)

axes.set_yscale("log")
axes.legend(loc = 'upper right', labelspacing = 0.05, fontsize = 10)


plt.savefig("4L_PL.png", bbox_inches='tight',dpi = 300)

plt.show()
```
