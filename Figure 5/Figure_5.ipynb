```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
import csv
import matplotlib.patches as mpatches
import pylab as pb
import matplotlib.ticker as mtick
from matplotlib import cm
from numpy import linspace
from csv import reader
from matplotlib import pyplot
import matplotlib.colors as colors
import bokeh as bk
```

```python
x1 = []
y1 = []


with open('fig_a_conductance.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x1.append(float(row[0]))
        y1.append(float(row[1]))
        
```

```python
x2 = []
y2 = []
z2 = []
a2 = []


with open('fig_b_concentration.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x2.append(float(row[0]))
        y2.append(float(row[1]))
        z2.append(float(row[2]))
        a2.append(float(row[3]))
        
```

```python
x3 = []
y3 = []


with open('fig_c_conductance_destroy.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x3.append(float(row[0]))
        y3.append(float(row[1]))
        
```

```python
x4 = []
y4 = []
z4 = []
a4 = []


with open('fig_d_concentration_destroy.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x4.append(float(row[0]))
        y4.append(float(row[1]))
        z4.append(float(row[2]))
        a4.append(float(row[3]))
```

```python
fig = plt.figure(figsize=(6,6))

fig.canvas.draw()

ax1 = fig.add_subplot(211)
ax1.plot(x1,y1,  ls = '-', lw = 2, color = 'blue', alpha=1, label = 'Simulated \nsheet conductance \n(conversion only)' )


ax2 = fig.add_subplot(212)
ax2.plot(x2,y2,  ls = '-', lw = 2, color = 'gold', alpha=1, label = 'MoS$_{2}$' )
ax2.plot(x2,z2, ls = '-', lw = 2, color = 'red', alpha=1, label = '2D-MoO$_{3}$' )
ax2.plot(x2,a2, ls = '-', lw = 2, color = 'green', alpha=1, label = 'MoO$_{3}$' )


ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")

ax2.set_xlabel('Iteration', fontsize = 12)
ax2.set_ylabel('Relative concentration', fontsize = 12)
ax1.set_ylabel('Relative conductance', fontsize = 12)
ax1.set_xlabel('None')
ax1.grid('on')
ax2.grid('on')

plt.setp(ax1.get_yticklabels(), visible=True, fontsize = 12)
plt.setp(ax2.get_yticklabels(), visible=True, fontsize = 12)
plt.setp(ax1.get_xticklabels(), visible=True, fontsize = 12)
plt.setp(ax2.get_xticklabels(), visible=True, fontsize = 12)



plt.setp(ax1.set_xlim([0,6200]))
plt.setp(ax2.set_xlim([0,6200]))

ax1.axvspan(1500, 2200, color='cadetblue', alpha=0.5)
ax2.axvspan(1500, 2200, color='cadetblue', alpha=0.5)

plt.setp(ax1.set_ylim([0,1.8]))
plt.setp(ax2.set_ylim([0,1.1]))


fig.subplots_adjust(hspace=0.0)
ax2.legend(loc='upper center',ncol = 3, fontsize = 12)
#ax2.legend(bbox_to_anchor=(0.9, 1.2),ncol = 2, fontsize = 12)
ax1.legend(loc='upper right',ncol = 1, fontsize = 12)


major_ticks_1 = np.arange(0, 1.81, 0.6)   
ax1.set_yticks(major_ticks_1)   

major_ticks_2 = np.arange(0, 1.01, 0.25)   
ax2.set_yticks(major_ticks_2)   



plt.savefig("Fig_5a_b.png",bbox_inches='tight', dpi = 300)


plt.show()
```
