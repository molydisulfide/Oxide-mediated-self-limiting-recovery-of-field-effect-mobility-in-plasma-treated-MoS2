```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp
import csv
import matplotlib.patches as mpatches
import pylab as pb
import matplotlib.ticker as mtick
from matplotlib import cm
from numpy import linspace
from csv import reader
from matplotlib import pyplot
import matplotlib.colors as colors
import bokeh as bk
```

```python
x5 = []
y5 = []



with open('Conductance_curve.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x5.append(float(row[0]))
        y5.append(float(row[1]))
      
```

```python
fig = plt.figure(figsize=(3,8))

fig.canvas.draw()

ax1 = fig.add_subplot(111)
ax1.plot(x5,y5,  ls = '-', lw = 6, color = 'crimson', alpha=1, label = 'Simulated \nsheet conductance' )
ax1.set_ylabel('Relative conductance', fontsize = 16)
ax1.set_xlabel('Exposure time (a.u.)', fontsize = 16)
ax1.grid('off')
plt.setp(ax1.get_yticklabels(), visible=True, fontsize = 16)
plt.setp(ax1.get_xticklabels(), visible=False, fontsize = 16)
plt.setp(ax1.set_xlim([0,2700]))
plt.setp(ax1.set_ylim([0,1.2]))
ax1.legend(loc='lower left',ncol = 1, fontsize = 10)
major_ticks_1 = np.arange(0, 1.21, 0.6)   
ax1.set_yticks(major_ticks_1)   
major_ticks_1 = np.arange(0, 1.21, 0.6)   
ax1.set_yticks(major_ticks_1)  
plt.savefig("Fig_5sim.png",bbox_inches='tight', dpi = 300)


plt.show()
```
